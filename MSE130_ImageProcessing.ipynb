```python
import numpy as np
import math as m
import cv2 as cv
import os
print(os.getcwd())
```

```python
#Opens the video file
cap = cv.VideoCapture('C:/Users/leoch_000/Desktop/MSE 130 Labs/Te15C-50X-9nm.mp4')
i = 0 # frame counter
j = 0 # time counter
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    print(i, i%42)
    if i%42 == 0: # Selects every ith frame
        cv.imwrite('Te15C'+str(j)+'s.jpg', frame) # Saves the frame as a .jpg archive
    i+=1 
    j = j+(1/2.78) # accounts for resolution of 2.78 frames per second
cap.release()
cv.destroyAllWindows()
```
